```python
# Problem 5: In the CENUS_DATA table list all community areas whose names start with the letter ‘B’.
```

```python
%%sql 
select COMMUNITY_AREA_NAME 
from CENSUS_DATA
where COMMUNITY_AREA_NAME like 'B%';
```

```python
# CHICAGO_PUBLIC_SCHOOLS
# Problem 6: List the schools in Community Areas 10 to 15 
# that are healthy school certified?
```

```python
%%sql 
select *
from CHICAGO_PUBLIC_SCHOOLS
where COMMUNITY_AREA_NUMBER >= 10
and COMMUNITY_AREA_NUMBER <= 15
and HEALTHY_SCHOOL_CERTIFIED = 'Yes';
```

```python
# Problem 7: What is the average school Safety Score?
```

```python
%%sql 
select AVG(SAFETY_SCORE) as AVG_SAFETY_SCORE
from CHICAGO_PUBLIC_SCHOOLS;
```

```python
# COLLEGE_ENROLLMENT
# Problem 8: List the top 5 Community Areas 
# by average College Enrollments [number of students]
```

```python
%%sql 
select COMMUNITY_AREA_NUMBER, COMMUNITY_AREA_NAME, 
    AVG(COLLEGE_ENROLLMENT) as AVG_COLLEGE_ENROLLMENT
from CHICAGO_PUBLIC_SCHOOLS
group by COMMUNITY_AREA_NUMBER, COMMUNITY_AREA_NAME
ORDER BY AVG_COLLEGE_ENROLLMENT DESC
LIMIT 5;
```

```python
# Problem 9: Use a sub-query to determine 
# which Community Area has the least value for Safety Score?
```

```python
%%sql
select COMMUNITY_AREA_NUMBER, COMMUNITY_AREA_NAME
from CHICAGO_PUBLIC_SCHOOLS
where SAFETY_SCORE = 
(select MIN(SAFETY_SCORE) from CHICAGO_PUBLIC_SCHOOLS);
```

```python
# Problem 10: Find the Per Capita Income of the Community Area 
# which has a school Safety Score of 1.
```

```python
%%sql
select A.COMMUNITY_AREA_NUMBER, 
       A.COMMUNITY_AREA_NAME,
       A.PER_CAPITA_INCOME
from CENSUS_DATA A, CHICAGO_PUBLIC_SCHOOLS B
where A.COMMUNITY_AREA_NUMBER = B.COMMUNITY_AREA_NUMBER
and B.SAFETY_SCORE = 1;
```
